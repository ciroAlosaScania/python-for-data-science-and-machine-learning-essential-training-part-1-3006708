# Introduction to Matplotlib and Seaborn

```python
# pip install seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame
```

```python
data = {'names': ['steve', 'john', 'richard', 'sarah', 'randy', 'michael', 'julie'],
        'age': [20, 22, 20, 21, 24, 23, 22],
        'gender': ['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female'],
        'rank': [2, 1, 4, 5, 3, 7, 6]}

df = DataFrame(data)
df
```

## Matplotlib's Bar Chart

```python
plt.bar(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

```python
# Bar plot with Seaborn it is more efficient, requires less code
plot = sns.barplot(data=df, x='names', y='age')
plot.set_title('Comparing Ages')
plot.set(xlabel='Names', ylabel='Age') # Not necessary, it already sets the names according to data
plt.show() # Use the matplotlib command to show the plot
```

## Line Plot using Matplotlib

```python
plt.plot(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

## Line Plot using Seaborn

```python
plot = sns.lineplot(data=df, x='names', y='age')
plot.set_title('Comparing Ages')
plt.show()
```

## PieChart using Matplotlib

```python
plt.pie(df['age'], labels=df['names'])
plt.title('Age Comparison')
plt.show()
```

## PieChart with Seaborn
Seaborn does not have a Pie Chart function, but the color scheme of Seaborn can be used on the Pie Chart from Matplotlib.

```python
color_scheme = sns.color_palette('pastel')[:5] # Take first 5 colors from the palette
plt.pie(df['age'], labels=df['names'], colors=color_scheme)
plt.show()
```

# Defining Elements of a Plot

In object-oriented plotting:
1. Create a blank figure object.
2. Add axes to the figure.
3. Generate plot(s) within the figure object.
4. Specify plotting and layout parameters for the plot(s) within the figure.


```python
# Tells python to print plot in line with notebook
%matplotlib inline 
```

```python
rcParams['figure.figsize'] = 5, 4  # 5*4 inches
```

## Defining axes, ticks and grids

*matplotlib.pyplot.figure.add_axes(rect)*

rect: This parameter is the dimensions [xmin, ymin, dx, dy] of the new axes. It takes the below elements as arguments in the list:

* xmin: Horizontal coordinate of the lower left corner.
* ymin: Vertical coordinate of the lower left corner.
* dx: Width of the subplot.
* dy: Height of the subplot.

Returns: This method return the axes class depends on the projection used. 



```python
x = range(1, 10)
y = [1, 2, 3, 4, 0, 4, 3, 2, 1]

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.plot(x, y)
```

### Add ticks and limits


```python
x = range(1, 10)
y = [1, 2, 3, 4, 0, 4, 3, 2, 1]

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.set_xlim([1, 9])
ax.set_ylim([0, 5])
ax.set_xticks([0, 1, 2, 3, 4, 5, 6, 8, 9, 10])
ax.set_yticks([0, 1, 2, 3, 4, 5])

ax.plot(x, y)
```

### Add a grid

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0, 4, 3, 2, 1]

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.grid()

ax.plot(x, y)
```

## Generating multiple plots in one figure with subplots
Create 1 figure with two plots.


```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1, 2) # rows and cols
ax1.plot(x)
ax2.plot(x, y)
```
